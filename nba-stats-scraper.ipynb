# Scrape Info From NBA Statistics Website for BUSN 265 Project 1

Statistics site: 2019-20 NBA Player Stats: Totals (https://www.basketball-reference.com/leagues/NBA_2020_totals.html)

```python
# import libraries
import requests
from bs4 import BeautifulSoup
```

```python
# save website to scrape in url
url = 'https://www.basketball-reference.com/leagues/NBA_2020_totals.html'
```

```python
# get page's HTML
page = requests.get(url)


# parse content using beautiful soup
soup = BeautifulSoup(page.content, 'html.parser')


# get tuple info of all players
rk = soup.select("table tbody tr.full_table th[data-stat=ranker]")
player = soup.select("table tbody tr.full_table td[data-stat=player] a")
pos = soup.select("table tbody tr.full_table td[data-stat=pos]")
age = soup.select("table tbody tr.full_table td[data-stat=age]")
tm = soup.select("table tbody tr.full_table td[data-stat=team_id] a")
g = soup.select("table tbody tr.full_table td[data-stat=g]")
gs = soup.select("table tbody tr.full_table td[data-stat=gs]")
mp = soup.select("table tbody tr.full_table td[data-stat=mp]")
fg = soup.select("table tbody tr.full_table td[data-stat=fg]")
fga = soup.select("table tbody tr.full_table td[data-stat=fga]")
fgp = soup.select("table tbody tr.full_table td[data-stat=fg_pct]")
three_pfg = soup.select("table tbody tr.full_table td[data-stat=fg3]")
three_pfga = soup.select("table tbody tr.full_table td[data-stat=fg3a]")
three_pfgp = soup.select("table tbody tr.full_table td[data-stat=fg3_pct]")
two_pfg = soup.select("table tbody tr.full_table td[data-stat=fg2]")
two_pfga = soup.select("table tbody tr.full_table td[data-stat=fg2a]")
two_pfgp = soup.select("table tbody tr.full_table td[data-stat=fg2_pct]")
efgp = soup.select("table tbody tr.full_table td[data-stat=efg_pct]")
ft = soup.select("table tbody tr.full_table td[data-stat=ft]")
fta = soup.select("table tbody tr.full_table td[data-stat=fta]")
ftp = soup.select("table tbody tr.full_table td[data-stat=ft_pct]")
orb = soup.select("table tbody tr.full_table td[data-stat=orb]")
drb = soup.select("table tbody tr.full_table td[data-stat=drb]")
trb = soup.select("table tbody tr.full_table td[data-stat=trb]")
ast = soup.select("table tbody tr.full_table td[data-stat=ast]")
stl = soup.select("table tbody tr.full_table td[data-stat=stl]")
blk = soup.select("table tbody tr.full_table td[data-stat=blk]")
tov = soup.select("table tbody tr.full_table td[data-stat=tov]")
pf = soup.select("table tbody tr.full_table td[data-stat=pf]")
pts = soup.select("table tbody tr.full_table td[data-stat=pts]")
```

```python
# print the length of each tuple item
print(f'length of rk: {len(rk)}')
print(f'length of player: {len(player)}')
print(f'length of pos: {len(pos)}')
print(f'length of age: {len(age)}')
print(f'length of tm: {len(tm)}')
print(f'length of g: {len(g)}')
print(f'length of gs: {len(gs)}')
print(f'length of mp: {len(mp)}')
print(f'length of fg: {len(fg)}')
print(f'length of fga: {len(fga)}')
print(f'length of fgp: {len(fgp)}')
print(f'length of three_pfg: {len(three_pfg)}')
print(f'length of three_pfga: {len(three_pfga)}')
print(f'length of three_pfgp: {len(three_pfgp)}')
print(f'length of two_pfg: {len(two_pfg)}')
print(f'length of two_pfga: {len(two_pfga)}')
print(f'length of two_pfgp: {len(two_pfgp)}')
print(f'length of efgp: {len(efgp)}')
print(f'length of ft: {len(ft)}')
print(f'length of fta: {len(fta)}')
print(f'length of ftp: {len(ftp)}')
print(f'length of orb: {len(orb)}')
print(f'length of drb: {len(drb)}')
print(f'length of trb: {len(trb)}')
print(f'length of ast: {len(ast)}')
print(f'length of stl: {len(stl)}')
print(f'length of blk: {len(blk)}')
print(f'length of tov: {len(tov)}')
print(f'length of pf: {len(pf)}')
print(f'length of pts: {len(pts)}')
```

```python
# using a loop, get each player's tuple info and store in a dictionary
players_info_list = []
for i in range(528):
    players_info_dict = {
        "rk" : rk[i].text,
        "player" : player[i].text,
        "pos" : pos[i].text,
        "age" : age[i].text,
        # "tm" : tm[i].text,  # has only 469 entries
        "g" : g[i].text,
        "gs" : gs[i].text,
        "mp" : mp[i].text,
        "fg" : fg[i].text,
        "fga" : fga[i].text,
        "fgp" : fgp[i].text,
        "three_pfg" : three_pfg[i].text,
        "three_pfga" : three_pfga[i].text,
        "three_pfgp" : three_pfgp[i].text,
        "two_pfg" : two_pfg[i].text,
        "two_pfga" : two_pfga[i].text,
        "two_pfgp" : two_pfgp[i].text,
        "efgp" : efgp[i].text,
        "ft" : ft[i].text,
        "fta" : fta[i].text,
        "ftp" : ftp[i].text,
        "orb" : orb[i].text,
        "drb" : drb[i].text,
        "trb" : trb[i].text,
        "ast" : ast[i].text,
        "stl" : stl[i].text,
        "blk" : blk[i].text,
        "tov" : tov[i].text,
        "pf" : pf[i].text,
        "pts" : pts[i].text,
    }
    players_info_list.append(players_info_dict)
```

```python
# store the scraped data to a csv file
import csv

# open file & create writer object
file = open('nba-player-stats-2019-2020.csv', mode='a', encoding='utf-8')
writer = csv.writer(file)

# write header to file
header = players_info_list[0].keys()
writer.writerow(header)

# write info to file
for info in players_info_list:
    writer.writerow(info.values())
```
